```python
ttt = []
for i in range(100):
    mu , kappa = 0 , 0
    r = np.random.vonmises(mu , kappa)/4
    r2 = np.random.vonmises(mu , kappa)/4
    ttt.append([r,r2])
ttt = np.array(ttt)
sns.kdeplot(ttt[:,0] , shade=True)
plt.title("von Mises distribution")
plt.show()
sns.kdeplot(ttt[:,0],ttt[:,1] , shade=True)
plt.show()
plt.scatter(ttt[:,0],ttt[:,1])
plt.show()
```

```python
ttt = []
for i in range(1000):
    mu , kappa = 0 , 0.5  # center , dispersion
    correction = np.pi
    cap_min, cap_max = 0, 1
    # von Mises in [-pi,pi]
    r = np.random.vonmises(mu , kappa)/(correction) + cap_max
    r2 = np.random.vonmises(mu , kappa)/(correction) + cap_max
    ttt.append([r/2,r2/2])
ttt = np.array(ttt)
sns.jointplot(ttt[:,0],ttt[:,1],kind="scatter")
plt.show()
```
